```python
# nbi:hide_in 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual, Layout
import ipywidgets as widgets
import random
from random import choices
from scipy import stats
from IPython.display import HTML
from scipy.stats import norm
import math

sns.set(font="serif", style = "white")

def ci4 (samplenum = 1000,samplesize = 500):
    std = 10      
    mu = 3
    sigma = std # standard deviation
    intervallist = []
    zsc = 1.96
    c = 0
    for i in range(0,samplenum):
        SampledFTN=np.random.normal(loc=loc, scale=std, size = 1000000)
        rs = random.choices(SampledFTN, k=samplesize)
        mean = np.mean(rs)
        upbound = mean + zsc*sigma/np.sqrt(samplesize)
        lowbound = mean - zsc*sigma/np.sqrt(samplesize)
        # collect difference between sample mean and mu
        intervallist.append([lowbound,mean,upbound])
        if upbound >= mu and lowbound <= mu:
            c += 1
        
    plt.figure(figsize=(20,10))
    plt.boxplot(intervallist, boxprops=dict(color="lightcoral", lw = 3))
    plt.plot([1, samplenum],[mu,mu], lw=4, color='darkorange', label = "Sample Size =" + str(samplesize))
    plt.plot([1, samplenum],[mu,mu], lw=4, color='firebrick', label = "True Mean")
    plt.legend(loc = 2 , fontsize = 18)
    plt.xlabel("The Samples", fontsize = 24)
    plt.ylim(0,6)
    plt.yticks(fontsize =18)
    plt.ylabel("The Value of the true Mean", fontsize = 24)
    plt.title("Confidence intervals of " + str(samplenum) + " samples from the population", fontsize = 18, color = "firebrick", y = 1.03)
    plt.xticks(color='w')
    plt.show()
    print("The Number of Intervals that cover the expected value is:", c)

style = {'description_width': 'initial'}    
int_widget = interactive(ci4,samplenum=widgets.IntSlider(value=0,min=100,max=1000,step=100, description = "Number of Samples taken from the population",readout_format='d',style = style),
                            samplesize=widgets.IntSlider(value=100,min=10,max=10000,step=10, description = "Sample Size",readout_format='d',style = style))
                    
int_widget.children[0].layout = Layout(width='700px')
int_widget.children[1].layout = Layout(width='700px')

int_widget.children[0].style.handle_color = 'firebrick'
int_widget.children[1].style.handle_color = 'darkorange'

display(int_widget)
     
display(HTML('''<style>[data-mime-type="application/vnd.jupyter.stdout"] pre { color:orange; font-size:24px}</style>'''))

```
